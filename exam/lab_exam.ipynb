```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
covid_impact = pd.read_csv("covid_impact.csv")
```

```python
covid_impact.info()
```

```python
covid_impact.head()
```

### Renaming 'Country/Region' and 'Provience/State' to 'Country' and 'States'.

```python
covid_impact.rename(columns={'Country/Region': 'Country'}, inplace=True)
covid_impact.rename(columns={'Provience/State': 'States'}, inplace=True)
```

### All the unique values in the 'Country' column.

```python
unique_countries = covid_impact['Country'].unique()
print(unique_countries)
```

### Number of times when the 'Deaths' were exactly 0.

```python
death_zero_count = (covid_impact['Deaths'] == 0).sum()
print(death_zero_count)
```

### Number of times when the 'Confirmed' cases were more that 1000.

```python
confirmed_1000_count = (covid_impact['Confirmed'] >= 1000).sum()
print(confirmed_1000_count)
```

### Mean 'Recovered' cases

```python
mean_recovered = covid_impact['Recovered'].mean()
print(mean_recovered)
```

### Standard deviation of 'Confirmed' cases?

```python
std_confirmed = covid_impact['Confirmed'].std()
print(std_confirmed)
```

### Variance of 'Deaths'.

```python
variance_deaths = covid_impact['Deaths'].var()
print(variance_deaths)
```

### Instances when 'Active' cases were recorded.

```python
active_cases = covid_impact[covid_impact['Active'] > 0]
print(active_cases)
```

### All instances when 'Confirmed' cases are above 1000 and 'Deaths' are 0.

```python
filtered_instances = covid_impact[(covid_impact['Confirmed'] > 1000) & (covid_impact['Deaths'] == 0)]
print(filtered_instances)
```

### Mean value of each column against each 'WHO Region'

```python
mean_by_who_region = covid_impact.groupby('WHO Region').mean()
print(mean_by_who_region)
```

### Minimum and maximum value of each column against each 'WHO Region'

```python
min_max_by_who_region = covid_impact.groupby('WHO Region').agg(['min', 'max'])
print(min_max_by_who_region)
```

### Instances when 'WHO Region' is 'Americas' or 'Confirmed' cases are above 5000.

```python
filtered_instances = covid_impact[(covid_impact['WHO Region'] == 'Americas') | (covid_impact['Confirmed'] > 5000)]
print(filtered_instances)
```

### Instances when 'WHO Region' is 'Americas' and 'Deaths' are greater than 50 or 'Recovered' is above 1000.

```python
filtered_instances = covid_impact[((covid_impact['WHO Region'] == 'Americas') & (covid_impact['Deaths'] > 50)) | (covid_impact['Recovered'] > 1000)]
print(filtered_instances)
```

### A bar chart showing the total number of confirmed cases for each country.

```python
confirmed_by_country = covid_impact.groupby('Country')['Confirmed'].sum()
confirmed_by_country.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.title('Total Confirmed Cases by Country')
plt.show()
```

### Line chart showing the trend of deaths over time globally.

```python
deaths_over_time = covid_impact.groupby('Date')['Deaths'].sum()
deaths_over_time.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('Global Death Trend Over Time')
plt.show()
```

### A pie chart showing the distribution of cases by WHO Region.

```python
cases_by_region = covid_impact.groupby('WHO Region')['Confirmed'].sum()
cases_by_region.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Distribution of Cases by WHO Region')
plt.show()
```

### A histogram showing the distribution of recovered cases.

```python
plt.hist(covid_impact['Recovered'], bins=10, edgecolor='black')
plt.xlabel('Recovered Cases')
plt.ylabel('Frequency')
plt.title('Distribution of Recovered Cases')
plt.show()
```

### A scatter plot showing the relationship between confirmed cases and deaths.

```python
plt.scatter(covid_impact['Confirmed'], covid_impact['Deaths'])
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.title('Relationship between Confirmed Cases and Deaths')
plt.show()
```

### A box plot showing the distribution of active cases.

```python
plt.boxplot(covid_impact['Active'])
plt.ylabel('Active Cases')
plt.title('Distribution of Active Cases')
plt.show()
```

### A bar chart showing the top 10 countries with the highest number of deaths.

```python
top_countries_deaths = covid_impact.groupby('Country')['Deaths'].sum().nlargest(10)
top_countries_deaths.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Total Deaths')
plt.title('Top 10 Countries with Highest Deaths')
plt.show()
```

### A line chart showing the trend of recovered cases over time globally.

```python
recovered_over_time = covid_impact.groupby('Date')['Recovered'].sum()
recovered_over_time.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Total Recovered Cases')
plt.title('Global Recovered Cases Trend Over Time')
plt.show()
```
